```python
import py3Dmol

def visualize_pdb(pdb_file):
    # Read the PDB structure
    with open(pdb_file, "r") as f:
        pdb_data = f.read()

    # Create 3Dmol viewer
    view = py3Dmol.view(js='https://3dmol.org/build/3Dmol.js', width=800, height=400)
    view.addModel(pdb_data, "pdb")  # Load the PDB content
    view.setStyle({'model': -1}, {"cartoon": {'color': 'spectrum'}})  # Cartoon style with spectrum coloring
    view.zoomTo()
    return view.show()
```

```python
for i in range(5):
    visualize_pdb(f"./pdbs/generated_protein_{i}.pdb")
```
